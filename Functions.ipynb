```python
create a function to print prime numbers with given range from user using function without input without return
```

```python
def prime_num():
    sv=int(input("enter a start value:"))
    sp=int(input("enter a stop value:"))
    for i in range(sv,sp+1,1):
        for j in range(2,i,1):
            if(i%j==0):
                break
        else:
            print(i,end=" ")
            
```

```python
prime_num()
```

```python
def prime(sv,sp):
    p1=[]
    for i in range(sv,sp+1,1):
        for j in range(2,i,1):
            if(i%j==0):
                break
        else:
            p1.append(i)
    return p1
```

```python
prime(2,50)
```

```python
lambda function
a lambda function is a small anonymous(nameless) function in python 

* It defined using keyword lambda instead of def 
* It can take any number of arguments but must continue only one expression
* Expression is automatically returned(no need to use return)

syntax
lambda argument:Expression
```

```python
sq=lambda x: x*x
sq(4)
```

```python
add = lambda a,b: a+b
add(2,3)
```

```python
Why to use lambda?
1) It will be useful for short simple operation
2) saves code(one-liner)
3) often used with map(),filter(),reduce() or as an inline function

when not use lambda?
1) for complex logic(use def instead)
2) when readbility is more important than shortness.
```

```python
create function to calculate factorial of given number using without input with return
```

```python
def fact():
    n=int(input("Enter number:"))
    factorial=1
    for i in range(1,n+1,1):
        factorial=factorial*i
    return n,factorial
fact()    
```

```python
v,r=fact()
print(f"the factorial of {v} is {r}") 
```
