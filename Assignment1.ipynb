```python
#Write a Python program to calculate area of square and circle.
r= float(input("Enter Radius of the circle:"))
Area = 3.14 *r*r
print(f"Area of circle is : {Area}")
```

```python
#Write a program that asks the usPer for the side of a square and prints its perimeter.
a = float(input("Enter side of square:"))
Perimeter= 4*a
print(f"Perimeter of square is :{Perimeter}")
```

```python
#Take the base and height of a triangle as input and print its area.
base = float(input("Enter base of triangle: "))
height = float(input("Enter height of triangle: "))
area = 0.5 * base * height
print(f"Area of Triangle : {area}")
```

```python
#Write a program that asks the user for the radius of a circle and prints its circumference. 
radius = float(input("Enter radius of circle: "))
circumference = 2 * 3.14 * radius
print("Circumference of Circle =", circumference)
```

```python
#5. Take Principal (P), Rate (R), and Time (T) as input from the user and print the Simple Interest.
P = float(input("Enter Principal: "))
R = float(input("Enter Rate: "))
T = float(input("Enter Time: "))
SI = (P * R * T) / 100
print(f"Simple Interest:{SI}")
```
