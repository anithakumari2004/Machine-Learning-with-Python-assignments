```python
nested if else syntax
if inside another if called as nested if 
if condition1: # outer if
    if condition2:# inner if
        statements cond2
    else:
        statements of inner else
else:
    statements of outer else 
```

```python
# check number is positive, negative or zero.
num = int(input("enter number:"))
if(num>=0):
    if(num==0):
        print("zero")
    else:
        print("Positive")
else:
    print("Negative")
```

```python
tb = float(input("enter total balance:"))
aw = float(input("Enter withdrawal amount"))  
if(tb>aw):
    if(aw%100==0):
        print("transaction successful")
    else:
        print("Enter amount in multiples of 100")
else:
    print("insufficient balance")
        
```

```python
discount calculator
>=5000 =10%
>=10000 = 20%
>=2000 is 5%
<2000 no discount
```

```python
purchase_amount = float(input("Enter purchase  amount:"))
if (purchase_amount>=5000):
    if (purchase_amount>=10000):
        print("Discount of 20%")
    else:
        print("Discount of 10%")
elif(purchase_amount>=2000):
    print("Discount of 5%")
else:
    print("no Discount")
```
